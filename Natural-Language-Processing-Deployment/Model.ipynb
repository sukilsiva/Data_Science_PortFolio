```python
### Importing the Libraries
import pandas as pd
import numpy as np
```

```python
### Read the Csv File
df = pd.read_csv("Data.csv", encoding='latin')
df.head()
```

1.In the Time Series Analysis we shouldnt split the data for train and test using train_test_split 

2.Instead Follow this method

```python
### Train and Test Split
train = df[df['Date'] < "2015-01-01"]
test = df[df["Date"] > "2014-12-31"]
print(len(train),len(test))
```

```python
### Pick the news data for feature engineering
data = train.iloc[:, 2:25]
data.head()
```

```python
### Rename the columns
list1 = [i for i in range(1,24)]
new_index = [str(x) for x in list1]
data.columns = new_index
data.head()
```

```python
### Remove all the Special Characters in dataset
data.replace("[^a-zA-Z]"," ",regex=True, inplace=True)
data.head()
```

```python
### All the words in dataset is getting lowerd to improve accuracy
for columns in new_index:
    data[columns] = data[columns].str.lower()
data.head()
```

```python
### Append all the data got transformed to headlines
headlines = []
for row in range(0, len(data.index)):
    headlines.append(" ".join(str(x) for x in data.iloc[row, 0:25]))
```

```python
headlines[0]
```

```python
### Import the Libraries for Model Prediction and Bag of Words
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_extraction.text import CountVectorizer
```

```python
### Creating the bag of words for train dataset
bag_of_words = CountVectorizer(ngram_range=(2, 2))
traindataset = bag_of_words.fit_transform(headlines)
```

```python
### Training the Model
classifier = RandomForestClassifier(n_estimators=200,criterion='entropy')
classifier.fit(traindataset,train['Label'])
```

```python
test_transform = []
for row in range(0, len(test.index)):
    test_transform.append(" ".join(str(x) for x in test.iloc[row, 2:27]))
test_transform[0]
```

```python
import re
corpus = []
for i in range(len(test_transform)):
    review = re.sub("[^a-zA-Z]"," ",test_transform[i])
    review = review.lower()
    review = review.split()
    corpus.append(" ".join(str(x) for x in review))
corpus[0]
```

```python
testdataset = bag_of_words.transform(corpus)
predictions = classifier.predict(testdataset)
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(test['Label'], predictions)
score
```

```python
### Pack the model for Deployment 
import pickle
pickle.dump(classifier, open("Classifier.pkl", "wb"))
pickle.dump(bag_of_words, open("CVtransform.pkl", "wb"))
```
